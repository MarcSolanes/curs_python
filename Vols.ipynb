```python
# importar llibreria pandas per poder llegir l'arxiu.

import pandas as pd
```

```python
# carregar el dataset al notebook

df_delays = pd.read_csv(r'/home/marc/Documents/DelayedFlights.csv')
```

```python
# Explorar el df. En primer lloc, busquem quines columnes conté

print(df_delays.columns.values)
```

```python
# Seguim amb l'exploració del df.

df_delays.describe()
```

```python
# Les columnes de retards contenen molts valors perduts (NaN). Suposarem que la falta de dades és perquè no es va 
# acumular un retard per aquell motiu. Per tant, substituirem els NaN per 0.

df_delays = df_delays.fillna(0)
```

```python
# Reduir el dataset a 200.000 files (en té casi 2 milions). El nou df per treballar s'anomena df_vols. Es fa servir el mètode de pandas "sample".

df_vols = df_delays.sample(n=200000)
```

```python
df_vols
```

## Exercici 2

```python
# Analitzar estadísticament el dataset. 

companyies = df_vols.UniqueCarrier.unique()

print(companyies)
print(len(companyies))
```

```python
print("El retard mitjà de cada vol va ser de " + str(round(df_vols.ArrDelay.mean(), 2)) + " minuts.")
```

```python
print("La duració mitjana de cada vol va ser de " + str(round(df_vols.AirTime.mean(), 2)) + " minuts.")
print("La desviació mitjana de la duració de vol va ser de  " + str(round(df_vols.AirTime.std(), 2)) + " minuts.")
print("Com que és un valor molt alt, pot ser interessant observar el valor de la mediana. La mediana de la duració de vol va ser de " + str(round(df_vols.AirTime.median(), 2)) + " minuts.")
```

```python
df_vols.describe()
```

### Observem com els retards més grans es donen per culpa de la companyia aèria i per l'arribada tardana de l'aparell. En ambdós casos el responsable n'és l'operador. 

```python
# Busquem quina aerolinia va tenir més retards de mitjana

retards_mitjans = df_vols.groupby("UniqueCarrier")["ArrDelay"].mean()
```

```python
# Els ordenem per veure el superior. 

retards_mitjans.sort_values(ascending = False)
```

```python
# Hi ha molta diferència entre els retards de B6 i YV i els de F9 i AQ. Podria ser per la distància dels seus vols. Comprove'm-ho.

distancies_mitjanes = df_vols.groupby("UniqueCarrier")["Distance"].mean()

distancies_mitjanes.sort_values(ascending = False)
```

### YV és la pitjor companyia aèria pel que fa als retards. Fa els vols de menor distància i, tot i així, és la segona companyia que més retards acumula. F9, a l'altre extrem, fa viatges entre mitjans i llargs i, en canvi, és la segona que menys retards acumula.

```python
# Quins són els viatges més llargs?

df_vols_llargs = df_vols.sort_values(by = ["Distance"], ascending = False)

df_vols_llargs.head(10)
```
