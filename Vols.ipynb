```python
# importar llibreria pandas per poder llegir l'arxiu.

import pandas as pd
```

```python
# carregar el dataset al notebook

df_delays = pd.read_csv(r'/home/marc/Documents/DelayedFlights.csv')
```

```python
# Explorar el df. En primer lloc, busquem quines columnes cont√©

print(df_delays.columns.values)
```

```python
# Seguim amb l'exploraci√≥ del df.

df_delays.describe()
```

```python
# Les columnes de retards contenen molts valors perduts (NaN). Suposarem que la falta de dades √©s perqu√® no es va 
# acumular un retard per aquell motiu. Per tant, substituirem els NaN per 0.

df_delays = df_delays.fillna(0)
```

```python
# Reduir el dataset a 200.000 files (en t√© casi 2 milions). El nou df per treballar s'anomena df_vols. Es fa servir el m√®tode de pandas "sample".

df_vols = df_delays.sample(n=200000)
```

```python
# Analitzar estad√≠sticament el dataset. 

companyies = df_vols.UniqueCarrier.unique()

print(companyies)
print(len(companyies))
```

### Observem com els retards m√©s grans es donen per culpa de la companyia a√®ria i per l'arribada tardana de l'aparell. En ambd√≥s casos el responsable n'√©s l'operador. 

```python
# Busquem quina aerolinia va tenir m√©s retards de mitjana

retards_mitjans = df_vols.groupby("UniqueCarrier")["ArrDelay"].mean()
```

```python
# Els ordenem per veure el superior. 

retards_mitjans.sort_values(ascending = False)
```

```python
# Hi ha molta difer√®ncia entre els retards de B6 i YV i els de F9 i AQ. Podria ser per la dist√†ncia dels seus vols. Comprove'm-ho.

distancies_mitjanes = df_vols.groupby("UniqueCarrier")["Distance"].mean()

distancies_mitjanes.sort_values(ascending = False)
```

### YV √©s la pitjor companyia a√®ria pel que fa als retards. Fa els vols de menor dist√†ncia i, tot i aix√≠, √©s la segona companyia que m√©s retards acumula. F9, a l'altre extrem, fa viatges entre mitjans i llargs i, en canvi, √©s la segona que menys retards acumula.

```python
# Quins s√≥n els viatges m√©s llargs?

df_vols_llargs = df_vols.sort_values(by = ["Distance"], ascending = False)

df_vols_llargs.head(10)
```
